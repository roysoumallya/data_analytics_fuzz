```python
df_sales = pd.concat([df_2022, df_2023], ignore_index=True)
```

```python
df_sales.tail()
```

```python
df_sales.head()
```

```python
df_data = df_sales.merge(df_product, on="Product_ID", how="left")
```

```python
df_data.head()
```

```python
df_data = df_data.merge(df_store, on="Store_ID", how="left")
```

```python
df_data.head()
```

```python
df_data['Product_Name / Category'].value_counts().head().plot(kind='bar')
```

```python
df_data['Date'] = pd.to_datetime(df_data['Date'], format="%d-%m-%Y")


df_data.head()
```

```python
df_data.shape
```

```python
df_data['Months'] = df_data['Date'].dt.month
```

```python
df_data.head()
```

```python
df_data.describe()
```

```python
df_data['Units'] = pd.to_numeric(
    df_data['Units'].str.replace(r'\s*Units?', '', regex=True),
    errors='coerce'
)
```

```python
df_data.head()
```

```python
df_data['Product_Price'] = df_data['Product_Price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df_data.head()
```

```python
df_data['Sales'] = df_data['Units']*df_data['Product_Price']
df_data[df_data["Units"]>1].head()
```

```python
df_data['Years'] = df_data['Date'].dt.year
```

```python
df_data[(df_data["Units"]>1) & (df_data["Years"]==2023)].head()
```

```python
monthly_sales = df_data.groupby(['Years', 'Months'])['Sales'].sum().reset_index()
monthly_sales

```

```python
plt.figure(figsize=(10,6))

sales_2022 = monthly_sales[monthly_sales['Years'] == 2022]
sales_2023 = monthly_sales[monthly_sales['Years'] == 2023]

plt.plot(sales_2022['Months'], sales_2022['Sales'], marker='o', label='2022')
plt.plot(sales_2023['Months'], sales_2023['Sales'], marker='o', label='2023')

plt.title("Monthly Sales Comparison: 2022 vs 2023")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.xticks(range(1, 13))
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_data.describe()
```

```python
total_sales_2022 = sales_2022['Sales'].sum()

print(total_sales_2022)
```

```python
df_data.groupby('Product_Name / Category')['Sales'].sum().head()
```

```python
df_data['Category'] = df_data['Product_Name / Category'].str.split(' - ').str[-1]
df_data.head()
```

```python
category_sales = (df_data.groupby(['Category', 'Years'])['Sales'].sum().reset_index())
```

```python
category_pivot = category_sales.pivot(
    index = 'Category',
    columns = 'Years',
    values = 'Sales').fillna(0)
print(category_pivot)
```

```python
category_pivot['Sales_Change'] = category_pivot[2023] - category_pivot[2022]
```

```python
declining_categories = category_pivot[category_pivot['Sales_Change'] < 0]
declining_categories.sort_values('Sales_Change')
```

```python
#Year-wise sales-change
```

```python
declining_categories['Sales_Change'].plot(kind='barh', title='Declining sales by Category')
plt.xlabel('Sales Diff')
plt.show()
```
